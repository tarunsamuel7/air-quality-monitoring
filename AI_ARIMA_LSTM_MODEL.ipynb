##### Calculating upper and lower fence of data

```python

def uf_lf(features):
    dic = {}
    for i in features:
        # calculating quartiles
        q25,q75 = data[i].quantile([0.25,0.75])

        # calculating Interquartile range
        IQR = q75 - q25

        # calculating lower and upper fence
        lf = q25 - 1.5*IQR
        uf = q75 + 1.5*IQR

        dic[i + '_lf']= lf
        dic[i + '_uf']= uf
    return dic

feat2 = ['Temp', 'Wind-Speed (U)', 'Dewpoint Temp', 'Soil Temp', 'Vegitation (High)', 'Vegitation (Low)', 'Total Percipitation', 'Relative Humidity', 'Wind-Speed (V)', 'NO2', 'O3', 'PM2.5','PM10' ]

uf_lf = uf_lf(feat2)
print(uf_lf)
```

```python
def calculate_outliers(data, uf_lf, features):
    for feature in features:
        outliers = data[feature][uf_lf[f'{feature}_uf'] < data[feature]].count() + data[feature][uf_lf[f'{feature}_lf'] > data[feature]].count()
        print(f'\n{feature} outliers count:', outliers)


# Example usage:
features = ['Temp', 'Wind-Speed (U)', 'Dewpoint Temp', 'Soil Temp', 'Vegitation (High)', 'Vegitation (Low)', 'Total Percipitation', 'Relative Humidity', 'Wind-Speed (V)', 'NO2', 'O3','PM10', 'PM2.5']
calculate_outliers(data, uf_lf, features)
```

####Handling Outiers

```python
# filling nan values with mean for fetaures with no or equal to no outliers
from scipy.stats import trim_mean
def fill_na(features):
    for feat in features:
        data[feat].fillna(data[feat].mean(), inplace=True)

features = ['Temp', 'Soil Temp', 'Relative Humidity', 'Dewpoint Temp', 'Wind-Speed (V)', 'O3' ]
fill_na(features)

# Filling missing values in Wind Speed (U)(using trim mean as data had outliers)
trim_mean_Wind = trim_mean(data['Wind-Speed (U)'], proportiontocut=0.05)
data['Wind-Speed (U)'].fillna(trim_mean_Wind, inplace=True)

# Filling missing values in Total Preception(using trim mean as data had outliers)
trim_mean_Preception = trim_mean(data['Total Percipitation'], proportiontocut=0.2)
data['Total Percipitation'].fillna(trim_mean_Preception, inplace=True)

# Filling missing values in NO2(using trim mean as data had outliers)
trim_mean_NO2 = trim_mean(data['NO2'], proportiontocut=0.1)
data['NO2'].fillna(trim_mean_NO2, inplace=True)

# Filling missing values in 'Vegitation (Low)'(using trim mean as data had outliers)
trim_mean_vegetation_low = trim_mean(data['Vegitation (Low)'], proportiontocut=0.1)
data['Vegitation (Low)'].fillna(trim_mean_vegetation_low, inplace=True)

# Filling missing values in 'Vegitation (High)'(using trim mean as data had outliers)
trim_mean_vegetation_High = trim_mean(data['Vegitation (High)'], proportiontocut=0.05)
data['Vegitation (High)'].fillna(trim_mean_vegetation_High, inplace=True)
```

```python
# Filling missing values in NO2(using median as data had outliers)
data['NO2'].fillna(data['NO2'].median(), inplace=True)

# Filling missing values in O3(using mean as data have very less outliers)
data['O3'].fillna(data['O3'].mean(), inplace=True)

# Filling missing values in PM2.5(using median as data had outliers)
data['PM2.5'].fillna(data['PM2.5'].median(), inplace=True)

# Filling missing values in PM10(using median as data had outliers)
data['PM10'].fillna(data['PM10'].median(), inplace=True)
```

```python
import itertools
import warnings
from statsmodels.tsa.stattools import adfuller

warnings.filterwarnings("ignore")

def optimize_arima(data, p_values, d_values, q_values):
    best_aic = float("inf")
    best_order = None
    best_model = None

    for p, d, q in itertools.product(p_values, d_values, q_values):
        try:
            model = ARIMA(data, order=(p, d, q)).fit()
            if model.aic < best_aic:
                best_aic = model.aic
                best_order = (p, d, q)
                best_model = model
        except:
            continue
    return best_order, best_model

# Define the range of parameters
p_values = range(0, 6)
d_values = range(0, 2)
q_values = range(0, 6)

# Optimize ARIMA parameters for NO2
best_order_no2, best_model_no2 = optimize_arima(no2_data, p_values, d_values, q_values)

best_order_no2, best_model_no2.aic
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Handle missing values
data.interpolate(method='linear', inplace=True)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data[['NO2', 'O3', 'PM2.5', 'PM10']])

# Create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 30  # Sequence length
X, y = create_sequences(scaled_data, seq_length)

# Split the data into training and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(seq_length, X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=4))  # Output layer with 4 units for NO2, O3, PM2.5, and PM10

model.compile(optimizer='adam', loss='mean_squared_error')


```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Build the LSTM model
model = Sequential()
model.add(Bidirectional(LSTM(units=64, return_sequences=True, input_shape=(seq_length, X_train.shape[2]))))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(units=64)))
model.add(Dropout(0.2))
model.add(Dense(units=4))  # Output layer with 4 units for NO2, O3, PM2.5, and PM10

model.compile(optimizer='adam', loss='mean_squared_error')

# Define early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_data=(X_test, y_test), callbacks=[early_stopping])

```

```python
# Evaluate the model
y_pred = model.predict(X_test)
y_test_inv = scaler.inverse_transform(y_test)
y_pred_inv = scaler.inverse_transform(y_pred)
```
